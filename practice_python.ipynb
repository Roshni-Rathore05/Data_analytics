```python
# string pelindrom
def is_pelindrom(s):
  n=len(s)
  left=0
  right=n-1
  while left<right:
    if s[left] != s[right]:
      print("not pelindrome")
      return False
    left +=1
    right -=1
  print("pelindrome")
  return True
print(is_pelindrom("abcdecba"))
```
